```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('dirty_cafe_sales.csv')
df.info()
```

```python
df.duplicated()
```

```python
missing_valuesSum = df.isnull().sum
```

```python
#replace all null known values as nan
junk_values = ['UNKNOWN', 'Unknown', 'unknown', 'ERROR', 'Error', 'error', '<NA>', 'na', 'inf', np.inf, 'NaN']
df = df.replace(junk_values, np.nan)
df
```

```python
#replace str objects to int
df['Quantity'] = df['Quantity'].astype('Float64')
df['Price Per Unit'] = df['Price Per Unit'].astype('Float64')
df['Total Spent'] = df['Total Spent'].astype('Float64')
```

```python
df= df.dropna(subset = ['Item'])
df = df.replace('', 0)
```

```python
df
```

```python
df['Total Spent'] = df['Total Spent'].fillna(0)
df['Quantity'] = df['Quantity'].fillna(0)
df['Price Per Unit'] = df['Price Per Unit'].fillna(0)

```

```python
df['Total Spent'] = np.where(df['Total Spent'] == 0, df['Quantity'] * df['Price Per Unit'], 
                             df['Total Spent'])
df['Quantity'] = np.where(df['Quantity'] == 0, df['Total Spent'] / df['Price Per Unit'].replace(0, 1), 
                             df['Total Spent'])

df['Price Per Unit'] = np.where(df['Price Per Unit'] == 0, df['Total Spent'] / df['Quantity'].replace(0, 1), df['Price Per Unit']) 
```

```python
df['Payment Method'] = np.where(
    (df['Payment Method'].isna()) & (df['Location'] == 'Online'), 
    'Credit Card', 
    df['Payment Method']
)
```

```python
df['Payment Method'] = df['Payment Method'].fillna('Cash')
```

```python
most_common_store = df['Location'].mode()[0]
df['Location'] = df['Location'].fillna(most_common_store)
```

```python
df = df.dropna(subset=['Transaction Date'])
```

```python
df = df.sort_values(by="Transaction Date", ascending = True)
```

```python

df['Transaction Date'] = df['Transaction Date'].ffill()
```

```python
df.to_csv('cleaned_cafe_sales.csv', index=False)
```
